### Hobbies vs. Caffeine Intake
<p> Our null hypothesis here is that the amopunt of time a person spends on hobbies and their caffeine intake are independent. Accoridng to the Pearson correlation test below, there is very little positive correlation between the number of hours a student spends on their hobby, and the amount of caffeine ingested daily. Despite the graph for this test suggesting that there may be a correlation, the Pairwise Correlation test concludes there is very little correlation after all. 

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

df = pd.read_csv("CS105.csv")
list1 = df['Hub Hours']
list2 = df['Intake']

list1 = np.nan_to_num(list1)
list2 = np.nan_to_num(list2)

corr, _ = pearsonr(list1, list2)
print("The Pearson COrrelation test concludes that the relationship between hobby hours and caffeine intake has a correlation score of %.3f" % corr)
```

### GPA vs. Caffeine Intake
<p> Below we have conducted a Chi-Squared test regarding a student's GPA and their caffeine intake. We want to test if there is any correlaction between a student's GPA and their daily caffeine intake. Our null hypothesis here is that there is no correlation between GPA and caffeine intake. Using the test we get the follwoing datapoints: </p>

- Chi-Squared: 40.836
- p-value: 0.75868
- Degrees of Freedom: 48

<p>When looking at the Chi-Squared table, we get a value of 65.171 for a 0.05 alpha value and 48 degrees of freedom. The Chi_squared value that we observed is much less than that, at a value 40.83 and the p-value is large at 0.758. While this would cause us to reject the null hypothesis, it is worth noting that with such a large p-value, our evidence from this data set may not be strong enough to prove our hypothesis. We beleive that there is to much variablility in this sample size to declare any correlation.

```python
import pandas as pd
from scipy.stats import chi2
from scipy.stats import chi2_contingency
data = pd.read_csv("CS105.csv")
bins = [-1, 1, 100, 200, float('inf')]
labels = ['0 mg', '1-100 mg', '101-200 mg', '200+ mg']

# Create a new column with the categorized values
data['Caffeine'] = pd.cut(data['Intake'], bins=bins, labels=labels)
# Create a table of GPA vs. Caffeine INTake
table = pd.crosstab(data['Caffeine'], data['UNITS'])
chi_Test, p, df, expected = chi2_contingency(table)

# critical value
significance_level = 0.05
godfrey_CV = chi2.isf(significance_level, df)

print(godfrey_CV)

# compare with the computed chi-squared statistic
if chi_Test >= godfrey_CV:
    print("Null hypothesis rejected") #meaning there is correlation? or am i upside ddown?
else:
    print("Null hypothesis not rejected") #meaning no correlation right? 

# Print the results
print("Chi-squared:")
print(f"Chi squared result: {chi_Test}")
print(f"p-value: {p}")
print(f"Degrees of freedom: {df}")
```
